```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from dax.models.dax_model1 import get_dax_forecasts_model1
```

```python
dax_forecasts_m1 = get_dax_forecasts_model1()
dax_forecasts_m1
```

```python
plt.plot(dax_forecasts_m1.index, dax_forecasts_m1.loc[:, 'q0.025':'q0.975'], ls="", marker="o", c="black")

plt.fill_between(dax_forecasts_m1.index, dax_forecasts_m1['q0.025'], dax_forecasts_m1['q0.975'],
                 color='gray', alpha=0.2, label='95% CI')
plt.fill_between(dax_forecasts_m1.index, dax_forecasts_m1['q0.25'], dax_forecasts_m1['q0.75'],
                 color='gray', alpha=0.4, label='50% CI')

plt.title('Dax Forecast Quantiles')
plt.ylabel('Log return')
plt.xlabel('Date')

# Show the plot
plt.show()
```
